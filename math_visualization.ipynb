```python
import numpy as np
import matplotlib.pyplot as plt
```

## Task1

```python
x = np.linspace(-10, 10, 200)
y1 = x
y2 = x**2
y3 = np.sin(x)
y4 = np.exp(-0.1*x) * np.cos(x)

plt.plot(x, y1, label='y = x')
plt.plot(x, y2, label='y = x²')
plt.plot(x, y3, label='y = sin(x)')
plt.plot(x, y4, label='y = e^(-0.1x) * cos(x)')

plt.title("Task1 - Function Plot")
plt.xlabel("x values")
plt.ylabel("y values")

plt.legend()
plt.grid()

plt.savefig("function_plot.png")

plt.show()
```

## Task2

```python
x = np.linspace(-20, 20, 400)

# Mixed equation using polynomial, sine, and absolute value terms
y = 0.01*x**4 + 0.1*x**3 - np.sin(x) - np.abs(x) + 3

plt.plot(x, y)

plt.title("Task2 - Own Equation")
plt.xlabel("x values")
plt.ylabel("y values")

plt.grid()

plt.savefig("own_equation.png")

plt.show()
```

## Task3

```python
students = ["S1", "S2", "S3", "S4", "S5", "S6", "S7", "S8", "S9", "S10"]
midterm = np.array([85, 72, 90, 66, 78, 92, 60, 74, 88, 95])
final = np.array([80, 70, 94, 68, 75, 90, 65, 72, 84, 96])
total = []
for i in range(len(midterm)):
    total.append(0.4 * midterm[i] + 0.6 * final[i])
```

```python
## scatter plot
plt.scatter(midterm, final)

plt.title("Task3 - Score Scatter Plot")
plt.xlabel("midterm score")
plt.ylabel("final score")

plt.grid()

plt.savefig('score_scatter.png')

plt.show()
```

```python
## Histogram
plt.hist(total)

plt.title("Task3 - Score Histrogram")
plt.xlabel("Score")
plt.ylabel("Students distribution")

plt.savefig('score_histogram.png')

plt.show()
```

```python
## Bar Chart
plt.bar(students, total)

plt.title("Task3 - Score Bar Chart")
plt.xlabel("Student")
plt.ylabel("Total Score")

plt.savefig('score_bar_chart.png')

plt.show()
```

## Task4

```python
slope, intercept = np.polyfit(midterm, final, 1)

final_pred = slope * midterm + intercept

plt.scatter(midterm, final, label="Original Data Points", alpha=0.5)

plt.plot(midterm, final_pred, label="Best-Fit Prediction line", color='red')

plt.title("Task4 - Score Prediction")
plt.xlabel("midterm score")
plt.ylabel("final score")

plt.legend()
plt.grid()

plt.savefig('score_prediction.png')

plt.show()


print("Prediction Examples")
example = np.array([50, 75, 100])
for score in example:
    print(f"midterm score = {score}, predicted final score = {slope * score + intercept}")
```
